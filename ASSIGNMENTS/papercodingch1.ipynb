Agency A is planning to issue tickets of a concert hall for idol singers' concert. Here, the number n is the input
and the seat number is arranged as follows. n * n seats are placed when n is given as input. The below
arrangement of the seat numbers is called a snake matrix because the array increases in ones in the shape of
a snake’s trunk. Write a program that produces arrays of these numbers.

```python
n = int(input("Enter the size of the matrix "))
++n
for i in range(0,n):
    x = i*n
    if(i%2 == 0):
         for j in range(x,x+n):
            print(j+1,end=" ")
    else:
        for j in range(x+n,x,-1):
            print(j,end=" ")
    print("")
```

A palindrome number refers to an integer whose value is the same as its original value, even if listed upside

The computer has a random integer between 1 and 100 as the correct answer value as following. When the
user presents the correct answer, the program only informs whether the presented integer is higher or lower
compared to the correct answer he or she stored. This game is repeated until the user answers correctly.

```python
import random
r = random.randint(1,100)
a = int(input("Guess a number between 1 and 100"))
while(a!=r):
    if a>r:
        print("Lower")
    if a<r:
        print("Higher")
    a = int(input("Guess again "))
print("Congratulations the correct number was ",r)
```
