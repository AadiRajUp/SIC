# 12 SNAKE MATRIX

# 15 reverse pattern

```python
x = int(input("Enter a number "))
b=x
p=0
valid = True
while(b!=0):
    b=b-p
    p=p+1
    if b<0:
        valid = False     
        print("Invalid input")
        break
if valid:
    for i in range (p-1,0,-1):
        for j in range (i):
            print(x,end=" ")
            x -= 1
        print("")
```

# 22 BMI

```python
height = int(input("Enter your height in cm: "))
weight = int(input("Enter your weight in kg: "))

height_in_m = height/100

bmi = weight/(height_in_m*height_in_m)

if(bmi < 18.5):
    print("Underweight")
elif(18.5 < bmi < 25):
    print("Normal weight")
elif(25 < bmi < 30):
    print("Pre-obesity ")
elif (30 < bmi < 35):
    print("Obesity class I")
elif (35 < bmi < 40):
    print("Obesity class II")
else:
    print("Obesity class III")
```
