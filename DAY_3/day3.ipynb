```python
valid = False
class AbsurdAge(Exception):
    pass
while(not valid):
    try:
        year = int(input("Enter your year of birth AD"))
        if year>2026 or year<1900:
            raise AbsurdAge("Invalid Age")
        else:
            year = 2026 - year
            break
    except AbsurdAge:
        print("Invalid year of birth",)
    except ValueError:
        print("Please enter a valid data")
print("Your age is ",year)
```

## WAP to take marks for 3 students as input form the user and give total score and average score

```python
average =0
total =0
for i in range(3):
    marks = int(input(f"Enter the marks of the student {i+1} "))
    total+=marks
average = total/3

print(f"The total marks is {total}, the average is {average}")
```
