```python
a = [1,2,3,1,1,1,2,3,4,3]
count = {}

for i in a:
    if i in count:
        count[i]+=1
    else:
        count[i] =1
print(count)
```

```python
a = [1,2,3,1,1,1,2,3,4,3]
count = {}

for i in a:
    count[i] = count.get(i,0) +1

count
```

```python
from collections import defaultdict
a = [1,2,3,1,1,1,2,3,4,3]

count = defaultdict(int)
for i in a: count[i]+=1
count
```
