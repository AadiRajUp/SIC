```python
for i in range(2,13):
    isPrime = True
    for j in range(2,i):
        if i!=j and i%j==0:
            isPrime = False
            break
        else:
            isPrime = True
    print(f"{i} is {"Prime" if isPrime else "Composite"}")
    
```
